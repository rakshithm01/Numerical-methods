# Fixed point iteration
## Question 1: Write a program to find the root of x=e^-2x using fixed point iteration


```python
import math
def g(x): # Defining the function to do the calculations
    return (math.e)**(-2*x)

err = 1e-10 # Defining the error margin of our choice
x0 = 0 # Initial guess value
count = 0 # To count the number of iterations it takes to find root
while(True): 
    x1 = g(x0)
    count = count + 1
    print('Iteration %0.f, x%0.f = %.11f, x%0.f = %.11f '% (count, count-1, x0, count, x1))
    if abs(x0-x1)<err: # Break condition when difference b/w successive solutions is less than the error
        break
    x0 = x1
print('The root of the equation is', x0, 'and the number of steps taken is', count)
```

# Newton-Raphson method
## Question 2: Write a program to find the root of any function using Newton-Raphson method

```python
x0 = 100 # Initial guess value
def f(x): # Defining a function to find its root
    return x**3+2*x+1

def fdash(x): # Derivative of the function
    return 3*x**2+2
  
count = 0
while(True):
    x1 = x0 - f(x0)/fdash(x0) # Iterative scheme for N-R methodd
    count = count + 1
    print('Iteration %0.f, x%0.f = %.11f, x%0.f = %.11f '% (count, count-1, x0, count, x1))
    if abs(x0-x1)<err:
        break
    x0 = x1
    
print('The root of the equation x\u00b3+2x+1 is', x0, 'and the number of steps taken is', count)
```

## As we can see, Newton-Raphson method is much faster than fixed point iteration. Only drawback is that we need to define the derivative of the function as well.
